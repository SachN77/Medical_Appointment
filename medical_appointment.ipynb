```python
import pandas as pd
import numpy as np

df = pd.read_csv("KaggleV2-May-2016.csv")
pd.set_option("display.max_columns", None)
df
```

```python
med_df = df[['Gender', 'Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'SMS_received','Handcap', 'No-show']]
med_df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

med_df['Gender'] = encoder.fit_transform(df['Gender'])
med_df['No-show'] = encoder.fit_transform(df['No-show'])
med_df
```

```python
x_columns = 8
X = med_df.iloc[:, 0: x_columns].values
y = med_df.iloc[:, x_columns].values
y
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
med_df = scale.fit_transform(med_df)
med_df
```

```python
from sklearn.model_selection import train_test_split
train_ratio = 0.80
test_ratio = 0.10
validation_ratio = 0.10

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_ratio)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = validation_ratio)


print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = "gini", splitter='best', max_leaf_nodes = 50000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
score = model.score(X_test, y_test)
print(score)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_RF = RandomForestClassifier(n_estimators = 1000, bootstrap = True)
model_RF.fit(X_train, y_train)

y_pred_RF = model_RF.predict(X_test)
score_RF = model.score(X_test, y_test)
print(score_RF)
```
